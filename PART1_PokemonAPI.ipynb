# 4 How many versions of Pokemon games have been released?

```python
url= 'https://pokeapi.co/api/v2/version-group'
response=requests.get(url)
data=response.json()
```

```python
version_stock=[]
i=len(data['results'])
for version_name in range(0,i):
    if data['results'][version_name]['name']not in version_stock:
            version_stock.append(data['results'][version_name]['name'])
print(len(version_stock))
```

# 6 What are electric-type Pokemon called in the Korean version of the game? (i.e. what do they call electric-type pokemon in Korean if not "electric")

```python
url= 'https://pokeapi.co/api/v2/type/13/'
response=requests.get(url)
data=response.json()
electric_url=[]
for a in data['pokemon']:
    electric_url.append(a['pokemon']['url'])
```

```python
for url in electric_url:
    response = requests.get(url)
    data = response.json()
    pokemon_url=data['species']['url']
    response2=requests.get(pokemon_url)
    data=response2.json()
    for a in data['names']:
        if a['language']['name']=='ko':
            print(a['name'])
```

# 7 Who has a higher speed stat, Eevee or Pikachu?

```python
url="https://pokeapi.co/api/v2/pokemon/eevee"
response = requests.get(url)
data = response.json()

for a in data['stats']:
    eevee_speed=a['base_stat']
```

```python
url="https://pokeapi.co/api/v2/pokemon/pikachu"
response = requests.get(url)
data = response.json()

for a in data['stats']:
    pikachu_speed=a['base_stat']
```

```python
if pikachu_speed>eevee_speed:
    print(f"Pikachu is faster")
else:
    print(f"Eevee is faster")
```
