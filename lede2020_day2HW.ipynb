```python
df2['Term[14]']=df2['Term[14]'].str.replace("Incumbent", datetime.now().strftime("%B %d, %Y"))
```

```python
import re
```

```python

date_pattern = r"([A-Za-z]+ \d{1,2}, \d{4})"

df2['Term[14]']=df2['Term[14]'].apply(lambda x: re.findall(date_pattern, x))

```

```python
date_format = "%B %d, %Y"


for date_range in df2['Term[14]']:
    
    start_date_str, end_date_str = date_range
    
    start_date = datetime.strptime(start_date_str, date_format).date()
    end_date = datetime.strptime(end_date_str, date_format).date()

    their_birthday="2010-01-01"
 
    their_birthday = datetime.strptime(their_birthday, "%Y-%m-%d").date()

    if start_date <= their_birthday <= end_date:
        a=df2[df2['Term[14]'].isin([date_range])]['Name (Birth–Death)']
        pattern = r"(\(\d{1,4}–\d{1,4}\)|\[\d+\])"
        print(re.sub(pattern, '',a).strip())

        
    else:
        pass

```

```python
correct_birthday = datetime.strptime("2010-01-01", "%Y-%m-%d").date()

# Apply date_pattern to df2['Term[14]'] to extract dates
df2['Term[14]'] = df2['Term[14]'].apply(lambda x: re.findall(date_pattern, ' '.join(x)))

# Iterate over each date range in df2['Term[14]']
for date_range in df2['Term[14]']:
    start_date_str, end_date_str = date_range
    
    # Parse start and end dates
    start_date = datetime.strptime(start_date_str, date_format).date()
    end_date = datetime.strptime(end_date_str, date_format).date()
    
    # Check if the correct_birthday falls within the date range
    if start_date <= correct_birthday <= end_date:
        # Filter rows based on the date_range
        filtered_rows = df2[df2['Term[14]'].apply(lambda x: x == date_range)]
        
        # Extract the name from the corresponding row
        name = filtered_rows['Name (Birth–Death)'].iloc[0]
        
        # Remove anything starting with '('
        name_without_parentheses = re.sub(r"\(.*", "", name)
        
        # Print the cleaned name
        print(name_without_parentheses.strip())
```
